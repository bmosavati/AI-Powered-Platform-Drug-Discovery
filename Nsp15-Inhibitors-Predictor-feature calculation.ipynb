```python
import numpy as np
import pandas as pd
from rdkit import Chem
from rdkit.Chem import Descriptors
import seaborn as sns
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
sol = pd.read_csv('SMILES-DATA - Equal-1 - Validation.csv')

# Define the function to calculate descriptors for a given SMILES string
def calculate_descriptors(smiles):
    mol = Chem.MolFromSmiles(smiles)
    if mol is None:
        return {key: None for key in descriptors.keys()}  # Return Nones if SMILES string is invalid

    descriptors = {
        'NHOHCount': Descriptors.NHOHCount(mol),
        'Chi3v': Descriptors.Chi3v(mol),
        'NumHDonors': Descriptors.NumHDonors(mol),
        'SlogP_VSA2': Descriptors.SlogP_VSA2(mol),
        'TPSA': Descriptors.TPSA(mol),
        'SlogP_VSA10': Descriptors.SlogP_VSA10(mol),
        'SMR_VSA2': Descriptors.SMR_VSA2(mol),
        'NOCount': Descriptors.NOCount(mol),
        'SlogP_VSA8': Descriptors.SlogP_VSA8(mol),
        'VSA_EState5': Descriptors.VSA_EState5(mol),
        'PEOE_VSA1': Descriptors.PEOE_VSA1(mol),
        'PEOE_VSA13': Descriptors.PEOE_VSA13(mol),
        'PEOE_VSA2': Descriptors.PEOE_VSA2(mol),
        'VSA_EState2': Descriptors.VSA_EState2(mol),
        'SlogP_VSA11': Descriptors.SlogP_VSA11(mol),
        'PEOE_VSA11': Descriptors.PEOE_VSA11(mol),
        'VSA_EState9': Descriptors.VSA_EState9(mol),
        'EState_VSA7': Descriptors.EState_VSA7(mol),
        'PEOE_VSA10': Descriptors.PEOE_VSA10(mol),
        'EState_VSA8': Descriptors.EState_VSA8(mol),
        'VSA_EState8': Descriptors.VSA_EState8(mol),
        'SlogP_VSA4': Descriptors.SlogP_VSA4(mol),
        'SMR_VSA7': Descriptors.SMR_VSA7(mol),
        'MolLogP': Descriptors.MolLogP(mol),
        'HallKierAlpha': Descriptors.HallKierAlpha(mol),
        'PEOE_VSA7': Descriptors.PEOE_VSA7(mol),
        'PEOE_VSA3': Descriptors.PEOE_VSA3(mol),
        'PEOE_VSA9': Descriptors.PEOE_VSA9(mol),
        'PEOE_VSA6': Descriptors.PEOE_VSA6(mol),
        'PEOE_VSA5': Descriptors.PEOE_VSA5(mol),
        'PEOE_VSA4': Descriptors.PEOE_VSA4(mol),
        'BalabanJ': Descriptors.BalabanJ(mol)
    }
    return descriptors



# Calculate descriptors for each molecule in the dataset
descriptor_list = []
for smiles in sol['SMILES']:
    descriptors = calculate_descriptors(smiles)
    descriptor_list.append(descriptors)
    
    print(descriptor_list)

# Convert descriptor list to DataFrame
descriptor_df = pd.DataFrame(descriptor_list)

# Compute the correlation matrix
correlation_matrix = descriptor_df.corr()


# Plot heatmap of the correlation matrix
plt.figure(figsize=(30, 20))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Chemical Descriptors', fontsize=55)

plt.xlabel('Descriptors')
plt.ylabel('Descriptors')
plt.xlabel('Descriptors', fontsize=20)  # Adjust X axis title fontsize
plt.ylabel('Descriptors', fontsize=20)  # Adjust Y axis title fontsize
plt.show()


# Save the heatmap as a PNG file
plt.savefig('correlation_heatmap.png', bbox_inches='tight', dpi=300)  # Adjust dpi for higher resolution if needed

# Save the heatmap as a PDF file
plt.savefig('correlation_heatmap.pdf', bbox_inches='tight')

# Close the current figure
plt.close()

# Now, show the heatmap
plt.show()
```

```python
import pandas as pd
from rdkit import Chem
from rdkit.Chem import Descriptors, Lipinski, MACCSkeys
from sklearn.preprocessing import MinMaxScaler

# Define the function to calculate new features (constitutional descriptors)
def calculate_constitutional_descriptors(smiles):
    mol = Chem.MolFromSmiles(smiles)
    if mol is None:
        return {key: None for key in descriptors.keys()}  # Return Nones if SMILES string is invalid
    
    descriptors = {
        'Molecular Weight': Descriptors.MolWt(mol),
        'LogP': Descriptors.MolLogP(mol),
        'Number of Atoms': mol.GetNumAtoms(),
        'Number of Bonds': mol.GetNumBonds(),
        'Number of Rings': Lipinski.RingCount(mol),
        'Rotatable Bonds Count': Lipinski.NumRotatableBonds(mol),
        'Hydrogen Bond Donors': Lipinski.NumHDonors(mol),
        'Hydrogen Bond Acceptors': Lipinski.NumHAcceptors(mol),
        'Number of Stereocenters': len(Chem.FindMolChiralCenters(mol, includeUnassigned=True)),
        'Topological Polar Surface Area (TPSA)': Descriptors.TPSA(mol),
    }
    return descriptors

# Calculate descriptors for each molecule
descriptors_list = [calculate_constitutional_descriptors(smiles) for smiles in sol['SMILES']]
descriptors_df = pd.DataFrame(descriptors_list)

# Normalize the descriptor values
#scaler = MinMaxScaler()
#normalized_descriptors_df = pd.DataFrame(scaler.fit_transform(descriptors_df.fillna(0)), columns=descriptors_df.columns)

# Save the normalized descriptors DataFrame to a CSV file
#normalized_descriptors_df.to_csv('normalized_descriptors-validation.csv', index=False)
descriptors_df.to_csv('descriptors-validation.csv', index=False)
```

```python
# Assuming `sol` is your DataFrame with the 'SMILES' column

# Define a function to convert SMILES strings to fingerprints
def smiles_to_fingerprint(smiles):
    mol = Chem.MolFromSmiles(smiles)
    fp = MACCSkeys.GenMACCSKeys(mol)
    fingerprint = np.array(list(fp.ToBitString()), dtype=int)
    return fingerprint

# Apply the function to the 'SMILES' column and create a new column for the fingerprints
sol['Fingerprint'] = sol['SMILES'].apply(smiles_to_fingerprint)

# Access the 'SMILES' and 'Fingerprint' columns
X = sol[['Fingerprint']]
X

sol['Fingerprint_Sum'] = sol['Fingerprint'].apply(lambda x: x.sum())

# Calculate the average of the 'Fingerprint_Sum' column
average_fingerprint_sum = sol['Fingerprint_Sum'].mean()

print(sol['Fingerprint_Sum'])
print(average_fingerprint_sum)
```
